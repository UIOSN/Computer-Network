```python
from huawei import switch1, switch2, HuaweiSwitch
from pathlib import Path
import time

log_dir = Path("log")
config_switch1 = Path("switch1_basic.txt")
config_switch2 = Path("switch2_basic.txt")
```

```python
with HuaweiSwitch(switch1, log_dir=log_dir) as S1:
    S1.sysname("SwitchA")
with HuaweiSwitch(switch2, log_dir=log_dir) as S2:
    S2.sysname("SwitchB")
```

步骤一

```python
# 接口镜像
mirror_cfg = [
    "observe-port interface GigabitEthernet 0/0/4",
    "interface GigabitEthernet0/0/1",
    "mirror to observe-port both",
    "quit",
    "interface GigabitEthernet0/0/2",
    "mirror to observe-port both",
    "quit",
    "interface GigabitEthernet0/0/3",
    "mirror to observe-port both",
    "quit"
]
with HuaweiSwitch(switch1, log_dir=log_dir) as S1:
    S1.send_commands(mirror_cfg)
```

```python
# (1) 查看两台交换机生成树的配置信息
with HuaweiSwitch(switch1, log_dir=log_dir) as S1:
    S1.send_command_y("undo stp enable")
    S1.send_commands("display stp")
```

```python
with HuaweiSwitch(switch2, log_dir=log_dir) as S2:
    S2.send_command_y("undo stp enable")
    S2.send_commands("display stp")
```

```python
# (3) ping 测试
# ping -t 192.168.1.20
# ping -t 192.168.1.30

# (4) 查看MAC地址表的变化
with HuaweiSwitch(switch1, log_dir=log_dir) as S1:
    for i in range(10):
        output = S1.send_commands("display mac-address")
        print(output)
        time.sleep(1)
```

步骤二

```python
with HuaweiSwitch(switch1, log_dir=log_dir) as S1:
    S1.configure(config_switch1)
```

步骤三

```python
with HuaweiSwitch(switch2, log_dir=log_dir) as S2:
    S2.configure(config_switch2)
```

步骤四

```python
with HuaweiSwitch(switch1, log_dir=log_dir) as S1:
    S1.send_command_y("stp enable")
    S1.send_commands("stp mode rstp")

with HuaweiSwitch(switch2, log_dir=log_dir) as S2:
    S2.send_command_y("stp enable")
    S2.send_commands("stp mode rstp")
```

```python
# ping -t 192.168.1.20
# ping -t 192.168.1.30

# 查看MAC地址表的变化
with HuaweiSwitch(switch1, log_dir=log_dir) as S1:
    for i in range(10):
        output = S1.send_commands("display mac-address")
        print(output)
        time.sleep(1)
```

步骤五

```python
# 先判断哪个为根交换机，再查看各端口的状态
with HuaweiSwitch(switch1, log_dir=log_dir) as S1:
    S1.send_commands("display stp interface GigabitEthernet0/0/1")
    S1.send_commands("display stp interface GigabitEthernet0/0/2")
```

```python
with HuaweiSwitch(switch2, log_dir=log_dir) as S2:
    S2.send_commands("display stp interface GigabitEthernet0/0/1")
    S2.send_commands("display stp interface GigabitEthernet0/0/2")
```
