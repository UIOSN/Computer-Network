```python
from huawei import switch1, switch2, HuaweiSwitch
from pathlib import Path
import time

log_dir = Path("log")
config_switch1 = Path("switch1_basic.txt")
config_switch2 = Path("switch2_basic.txt")
```

步骤一

```python
# 接口镜像
mirror_cfg = [
    "observe-port 1 interface GigabitEthernet 0/0/4",
    "interface GigabitEthernet0/0/1",
    "port-mirroring to observe-port 1 both",
    "quit",
    "interface GigabitEthernet0/0/2",
    "port-mirroring to observe-port 1 both",
    "quit",
    "interface GigabitEthernet0/0/3",
    "port-mirroring to observe-port 1 both",
    "quit"
]
with HuaweiSwitch(switch1, log_dir=log_dir) as S1:
    S1.send_commands(mirror_cfg)
```

```python
# (1) 查看两台交换机生成树的配置信息
with HuaweiSwitch(switch1, log_dir=log_dir) as S1:
    S1.send_command_y("undo stp enable")
    S1.send_commands("display stp")
```

```python
with HuaweiSwitch(switch2, log_dir=log_dir) as S2:
    S2.send_command_y("undo stp enable")
    S2.send_commands("display stp")
```

```python
# (3) ping 测试
# ping -t 192.168.1.20
# ping -t 192.168.1.30

# (4) 查看MAC地址表的变化
with HuaweiSwitch(switch1, log_dir=log_dir) as S1:
    for i in range(10):
        output = S1.send_commands("display mac-address")
        print(output)
        time.sleep(1)
```
